## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA

Remove the outliers

1

```python
sns.boxplot(x=df['Number_of_Vehicles'])
plt.title('Box Plot')
plt.xlabel('value')
plt.show()
```

2

```python
Q1 = df['Number_of_Vehicles'].quantile(0.25)
Q3 = df['Number_of_Vehicles'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Number_of_Vehicles'] < (Q1 - 1.5 * IQR)) | (df['Number_of_Vehicles'] > (Q3 + 1.5 * IQR)))]
```

Impute missing data

```python
num_df = df.select_dtypes(include =['number'])
num_df.fillna(num_df.mean(), inplace=True)
df[num_df.columns] =num_df
```

Scale the data

```python
df.dtypes
```

```python
scaler = StandardScaler()
numerical_features = ['Location_Easting_OSGR', 'Location_Northing_OSGR',
                      'Accident_Severity','Longitude', 'Latitude',
                      'Number_of_Vehicles','Number_of_Casualties','Year']
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

Reduce dimentions

```python
pca = PCA(n_components= 0.95)
pca_result = pca.fit_transform(df[numerical_features])
explained_variance_ratio = pca.explained_variance_ratio_
num_components =len(explained_variance_ratio)

print(explained_variance_ratio)
print(num_components)
```

```python
pca = PCA(n_components=0.95)
pca_result = pca.fit_transform(df[numerical_features])
df_pca = pd.DataFrame(data=pca_result, columns=['PC!', 'PC2' , 'PC3' , 'PC4', 'PC5','PC6'])
```

One-Hot Encoding

```python
df = pd.get_dummies(df, columns=['Weather_Conditions'])
```
